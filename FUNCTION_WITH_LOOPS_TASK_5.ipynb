```python
#1.Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
  total_likes=sum(likes_list)
  print("Total Likes:",total_likes)
  if total_likes>=1000:
    print("Post Status: Viral Post")
  else:
    print("Post Status: Normal Engagement")
likes_data=[200,150,300,400]
analyze_post_engagement(likes_data)
```

```python
#2.Healthcare – Medicine Stock Alert System
def check_medicine_stock(stock):
  print("Medicine Stock:",stock)
  if stock<10:
    print("Status: Low Stock Alert")
  else:
    print("Status: Stock Sufficient")
check_medicine_stock(6)
```

```python
#3.Agriculture – Rainfall Adequacy Checker
def check_rainfall_adequacy(rainfall_list,required_level):
  total_rainfall=sum(rainfall_list)
  average_rainfall=total_rainfall/len(rainfall_list)
  print("Average Rainfall:",average_rainfall)
  if average_rainfall>=required_level:
    print("Rainfall Status: Adequate Rainfall")
  else:
    print("Rainfall Status: Inadequate Rainfall")
seasonal_data=[70,75,80,63]
required=70
check_rainfall_adequacy(seasonal_data,required)
```

```python
#4.Social Media – Duplicate Account Detection
def check_duplicate_account(usernames):
  unique_users=set(usernames)
  if len(unique_users)<len(usernames):
    print("Duplicate Accounts Found: Yes")
  else:
    print("Duplicate Accounts Found: No")
user_list=["123","456","789","123"]
check_duplicate_account(user_list)
```

```python
#5.Healthcare – Appointment Eligibility Checker
def check_appointment_eligibility(age):
  print("Patient Age:",age)
  if age>=18:
    print("Eligibility Status: Eligible")
  else:
    print("Eligibility Status: Not Eligible")
check_appointment_eligibility(21)
```

```python
#6.Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
  premium_list=[]
  for price in prices:
    if price>2000:
      premium_list.append(price)
  print("Premium Crops:",premium_list)
crop_prices=[1500,2500,1800,3200]
filter_premium_crops(crop_prices)
```

```python
#7.System Monitoring – Application Health Checker
def check_application_health(error_count):
  print("Error Count:",error_count)
  if error_count==0:
    print("System Status: Healthy")
  elif error_count<=5:
    print("System Status: Minor Issues")
  else:
    print("System Status: Critical Issues")
check_application_health(7)
```

```python
#8.Banking – Daily Transaction Limit Checker
def check_transaction_limit(amount):
  daily_limit=50000
  print("Transaction Amount:",amount)
  if amount<=daily_limit:
    print("Transaction Status: Approved")
  else:
    print("Transaction Status: Rejected")
check_transaction_limit(60000)
```

```python
#9.E-Learning – Student Attendance Eligibility System
def check_exam_eligibility(attendance_list):
  total_classes=len(attendance_list)
  attended_classes=0
  for status in attendance_list:
    if status==1:
      attended_classes+=1
  attended_percentage=(attended_classes/total_classes)*100
  print("Attendance Percentage:",attended_percentage)
  if attended_percentage>=75:
    print("Exam Eligibility: Eligible")
  else:
    print("Exam Eligibility: Not Eligible")
student_attendance=[1,1,1,1,0]
check_exam_eligibility(student_attendance)
```

```python
#10.Smart Electricity Bill Analyzer
def calculate_electricity_bill(units):
  bill=0
  if units<=100:
    bill=units*3
  elif units<=200:
    bill=(100*3)+(100*5)+(units-200)*7
  else:
    bill=(100*3)+(100*5)+(units-200)*7
  if bill<500:
    usage_status="Low Usage"
  elif bill<1500:
    usage_status="Moderate Usage"
  else:
    usage_status="High Usage"
  return bill,usage_status
units_consumed=250
total_bill,status=calculate_electricity_bill(units_consumed)
print("Units Consumed",units_consumed)
print("Total Bill:",total_bill)
print("Usage Classification:",status)
```
