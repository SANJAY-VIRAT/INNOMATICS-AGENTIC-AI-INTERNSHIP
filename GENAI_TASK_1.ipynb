Import

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

Task 1 – Basic PromptTemplate

```python
template1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print(template1.format(topic="Artificial Intelligence"))
```

Task 2 – Multi Input Template

```python
template2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print(template2.format(topic="AI", audience="beginners", tone="friendly"))
print(template2.format(topic="Python", audience="kids", tone="fun"))
print(template2.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

Task 3 – Prompt Variations

```python
teaching = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

storytelling = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching.format(topic=topic))
print(interview.format(topic=topic))
print(storytelling.format(topic=topic))
```

Task 4 – ChatPromptTemplate (Roles)

```python
role_map = {
    "teacher": "You are a helpful teacher who explains concepts clearly.",
    "interviewer": "You are an interviewer who asks insightful questions.",
    "motivator": "You motivate and inspire learning."
}

role = "teacher"
topic = "Neural Networks"

chat_template = ChatPromptTemplate.from_messages([
    ("system", role_map.get(role)),
    ("user", "Explain {topic}")
])

messages = chat_template.format_messages(topic=topic)

print(messages)
```

Task 5 – Input Validation

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone


# Test
print(validate_inputs("kids", "fun"))
print(validate_inputs("expert", "cool"))
```

Task 6 – Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    style_templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions on {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=style_templates.get(style, style_templates["teaching"])
    )

    return template.format(topic=topic, audience=audience, tone=tone)


# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

Task 7 – Reusability Test

```python
template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "expert", "formal"),
    ("ML", "intermediate", "casual"),
    ("DL", "beginner", "formal"),
    ("Data Science", "expert", "fun")
]

for topic, audience, tone in inputs:
    print(template.format(topic=topic, audience=audience, tone=tone))
```
