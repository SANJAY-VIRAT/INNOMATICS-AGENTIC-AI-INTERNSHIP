```python
#1.Smart Parking Lot Management System
def parking_lot_management_system(total_capacity,logs):
    currently_parked=0
    peak_usage=0
    for log in logs:
      if log=="IN":
          currently_parked+=1
      elif log=="OUT":
        if currently_parked>0:
          currently_parked-=1
      if currently_parked>peak_usage:
        peak_usage=currently_parked
    return currently_parked,peak_usage
total_capacity=50
logs=["IN","IN","IN","OUT","IN","IN","OUT"]
currently_parked,peak=parking_lot_management_system(total_capacity,logs)
print("Currently Parked Vehicles:",currently_parked)
print("Peak Parking Usage:",peak)
if currently_parked>total_capacity:
  print("Parking Status: Available")
else:
  print("Parking Status: Not Available")
```

```python
#2.Online Food Delivery Time Estimator
def estimate_delivery_time(distance,traffic,weather):
    base_time=distance*5
    if traffic=="Low":
        traffic_delay=5
    elif traffic=="Medium":
        traffic_delay=10
    elif traffic=="High":
        traffic_delay=15
    else:
        traffic_delay=0
    if weather=="Clear":
        weather_delay=0
    elif weather=="Cloudy":
        weather_delay=5
    elif weather=="Rainy":
        weather_delay=10
    else:
        weather_delay=0
    total_time=base_time+traffic_delay+weather_delay
    return total_time
distance=8
traffic="High"
weather="Rainy"
eta=estimate_delivery_time(distance,traffic,weather)
print("Estimated Delivery Time:",eta,"minutes")
```

```python
#3.Movie Theatre Seat Occupancy Analyzer
def analyze_seat_occupancy(total_seats,booked_seats):
  booked_count=0
  for seat in booked_seats:
    if seat==1:
      booked_count+=1
  occupancy=(booked_count/total_seats)*100
  print("Occupancy:",occupancy,"%")
  if occupancy==100:
    print("Show Status: House Full")
  elif occupancy>=75:
    print("Show Status: Almost Full")
  else:
    print("Show Status: Seats Available")
total_seats=200
booked_seats=[1]*150
occupancy=analyze_seat_occupancy(total_seats,booked_seats)
```

```python
#4.Cloud Server Load Classification System
def classify_load(cpu_readings):
  total_cpu=0
  total_cpu_load=sum(cpu_readings)
  average_cpu_load=total_cpu_load/len(cpu_readings)
  print("Average CPU Load:",average_cpu_load)
  if average_cpu_load>80:
    print("Server Status: Critical")
  elif average_cpu_load>=50 and average_cpu_load<=80:
    print("Server Status: Warning")
  else:
    print("Server Status: Normal")
cpu_readings=[45,60,70,85,90]
classify_load(cpu_readings)
```

```python
#5.Smart Classroom Resource Usage Monitor
def monitor_resource_usage(resource_usage):
    threshold=8
    overused=[]
    for resource, usage in resource_usage.items():
        if usage>threshold:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resource_usage={
 "Projector": 6,
   "AC": 9,
 "Lights": 4
}
max_usage=monitor_resource_usage(resource_usage)
```

```python
#6.Online Event Registration Capacity Controller
def control_capacity(capacity,registrations):
  confirmed=0
  waitlisted=0
  for i in range(registrations):
    if confirmed<capacity:
      confirmed+=1
    else:
      waitlisted+=1
  print("Confirmed Registrations:",confirmed)
  if waitlisted>0:
    print("Waitlisted Users:",waitlisted)
    print("Registration Status: Closed")
  else:
    print("Waitlisted Users:",0)
    print("Registration Status: Open")
event_capacity=100
registrations=105
control_capacity(event_capacity,registrations)
```
