```python
#1.Employee Performance Bonus Eligibility
employees={}
n=int(input("Enter the number of employees: "))
for i in range(n):
    name=input("Enter employee name: ")
    score=int(input("Enter performance score: "))
    employees[name]=score
if n==0:
    print("No employees entered.")
else:
    highest_score=max(employees.values())
    top_employees=[]
    for name in employees:
        if employees[name]==highest_score:
            top_employees.append(name)
    print(", ".join(top_employees)+f" (Score: {highest_score})")
```

```python
#2.Search Query Keyword Analysis
s=input().lower()
s=s.split()
freq={}
for i in s:
  if i not in freq:
    freq[i]=1
  else:
    freq[i]+=1
filtered_freq={}
for i,j in freq.items():
  if j>=2:
    filtered_freq[i]=j
print(filtered_freq)
```

```python
#3.Sensor Data Validation
sensor_readings=list(map(int,input().split()))
res=[]
for i in range(len(sensor_readings)):
  if sensor_readings[i]%2==0:
    res.append((i,sensor_readings[i]))
print("Valid Sensor Readings (Hour, Value):")
print(res)
```

```python
#4.Email Domain Usage Analysis
emails=list(map(str,input().split()))
gmail_count=0
yahoo_count=0
for i in emails:
  if i[len(i)-10:]=="@gmail.com":
    gmail_count+=1
  elif i[len(i)-10:]=="@yahoo.com":
    yahoo_count+=1
gp=(gmail_count/len(emails))*100
yp=(yahoo_count/len(emails))*100
print(f"gmail.com: {gp:.0f}%")
print(f"yahoo.com: {yp:.0f}%")
```

```python
#5.Sales Spike Detection
sales=list(map(int,input().split()))
sales_avg=sum(sales)//len(sales)
sales_threshold=sales_avg*1.30
for i in range(len(sales)):
  if sales[i]>sales_threshold:
    print(f"Day {i+1}: {sales[i]}")
```

```python
#6.Duplicate User ID Detection
user_ids=list(map(str,input().split()))
freq={}
for i in user_ids:
  if i not in freq:
    freq[i]=1
  else:
    freq[i]+=1
for i, j in freq.items():
  if j>1:
    print(i,"->",j,"times")
```
