```python
#1.Smart Light Controller
class smart_light:
  def __init__(self,name):
    self.name=name
    self.status="OFF"
  def turn_on(self):
    self.status="ON"
  def turn_off(self):
    self.status="OFF"
  def show_status(self):
    print(f"{self.name} is {self.status}")
name=input("Light Name: ")
action=input("Action (ON/OFF): ")
light=smart_light(name)
if action=="ON":
  light.turn_on()
elif action=="OFF":
  light.turn_off()
light.show_status()
```

```python
#2.Employee ID Card System
class Employee:
  def __init__(self,name,id,dept):
    self.name=name
    self.id=id
    self.dept=dept
  def display_details(self):
    print("Employee ID Card")
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    print(f"Department: {self.dept}")
name=input("Employee Name: ")
id=input("Employee ID: ")
dept=input("Department: ")
employee=Employee(name,id,dept)
employee.display_details()
```

```python
#3.Mobile Contact Record
class Mobile_Contact:
  def __init__(self,name,number):
    self.name=name
    self.number=number
  def display_details(self):
    print("Contact Saved")
    print(f"Name: {self.name}")
    print(f"Phone: {self.number}")
name=input("Contact Name: ")
number=input("Phone Number: ")
contact=Mobile_Contact(name,number)
contact.display_details()
```

```python
#4.Product Price Tag Generator
class Product:
  def __init__(self,name,price):
    self.name=name
    self.price=price
  def display_details(self):
    print(f"Product: {self.name}")
    print(f"Price: {self.price}")
name=input("Product Name: ")
price=input("Price: ")
product=Product(name,price)
product.display_details()
```

```python
#5.Movie Rating Display System
class Movie:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating
  def display_details(self):
    print(f"Movie: {self.name}")
    print(f"Rating: {self.rating} / 5")
name=input("Movie Name: ")
rating=input("Rating: ")
movie=Movie(name,rating)
movie.display_details()
```

```python
#6.Delivery Address Manager
class Delivery_Address:
  def __init__(self,name,address):
    self.name=name
    self.address=address
  def display_details(self):
    print("Delivery Details")
    print(f"Customer: {self.name}")
    print(f"Address: {self.address}")
name=input("Customer Name: ")
address=input("Address: ")
address_book=Delivery_Address(name,address)
address_book.display_details()
```
