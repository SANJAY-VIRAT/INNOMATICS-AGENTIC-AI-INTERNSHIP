```python
#1.User Login Check
original_username="admin"
original_password="1234"
username=input("enter username:")
password=input("enter password:")
if original_username==username and original_password==password:
    print("Login Successful")
else:
    print("Invalid credentials")
```

```python
#2.Pass/Fail Analyzer
marks=[45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for i in marks:
  if i>=50:
    pass_count+=1
  else:
    fail_count+=1
print("Pass Count:",pass_count)
print("Fail Count:",fail_count)
```

```python
#3.Simple Data Cleaner
names=[" Alice ", "bob", " CHARLIE "]
res=[]
for i in names:
  i=i.strip()
  i=i.lower()
  res.append(i)
print(res)
```

```python
#4.Message Length Analyzer
messages=["Hi", "Welcome to the platform", "OK"]
for i in messages:
  if len(i)>10:
    print(f"{i}: This message contains more than 10 characters")
```

```python
#5.Error Message Detector
logs=["INFO", "ERROR", "WARNING", "ERROR"]
error_count=0
for i in logs:
  if i=="ERROR":
    error_count+=1
print(error_count)
```
