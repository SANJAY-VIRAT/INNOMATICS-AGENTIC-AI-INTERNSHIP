```python
#1.Mobile Recharge Validation System
def recharge_validation(amount):
  valid_plans=[199,299,399,599]
  if amount<50:
    print("Recharge amount must be at least 50 rupees")
    return False
  if amount not in valid_plans:
    print("Invalid recharge plan")
    print("Valid recharge plans are: 199,299,399,599")
    return False
  return True
while True:
  amount=int(input("Enter recharge amount: "))
  if recharge_validation(amount):
    print("Recharge successful!")
    break
  else:
    print("Please try again.")
```

```python
#2.Inventory Reorder Alert System
def check_inventory(products):
  for product,stock in products.items():
    if stock<15:
      print(f"{product} stock is low. Reorder now! Stock: {stock}")
    else:
      print(f"{product} stock Ok Stock: {stock}")
inventory={
    "Rice":25,
    "Sugar":10,
    "Oil":5,
    "Wheat":18,
    "Salt":30
}
check_inventory(inventory)
```

```python
#3.Student Result Processing System
def process_result(mark_list):
  total=0
  for mark in mark_list:
    total+=mark
  average=total/len(mark_list)
  print(f"Average Marks: {average:.2f}")
  if average>=50:
    return "Pass"
  else:
    return "Fail"
  marks=[60,45,70,55,40]
  result=process_result(marks)
  print("Result:",result)
```

```python
#4.Cab Fare Estimator with Retry Option
def calculate_fare(distance,peak):
  base_fare=50
  per_km_rate=12
  total=base_fare+(distance*per_km_rate)
  if peak.lower()=="yes":
    total=total+(0.25*total)
  return total
while True:
  distance=float(input("Enter distance in km: "))
  peak=input("Is it a peak hour? (yes/no): ")
  fare=calculate_fare(distance,peak)
  print(f"Total Fare: {fare:2f}")
  retry=input("Do you want to calculate another fare? (yes/no): ")
  if retry.lower()!="yes":
    print("Thank you for using Cab Fare Estimator!")
    break
  print()
```

```python
#5.Employee Attendance Eligibility Checker
def check_eligibility(attendance_list):
  present_count=0
  for status in attendance_list:
    if status=="P":
      present_count+=1
  total_days=len(attendance_list)
  attendance_percentage=(present_count/total_days)*100
  print(f"Attendance Percentage: {attendance_percentage:.2f}%")
  if attendance_percentage>=75:
    return "Eligible"
  else:
    return "Not Eligible"
attendance=["P","A","P","A","P"]
result=check_eligibility(attendance)
print("Eligibility Status:",result)
```

```python
#6.Password Strength Checker
def check_password_strength(password):
  has_digit=False
  has_special=False
  special_chars="@#$"
  if len(password)<8:
    return "Password must contain atleast 8 characters"
  for ch in password:
    if ch.isdigit():
      has_digit=True
    if ch in special_chars:
      has_special=True
  if has_digit and has_special:
    return "Password is strong"
  else:
    return "Password must contain at least one digit and one special character"
pwd=input("Enter Password:")
result=check_password_strength(pwd)
print(result)
```
