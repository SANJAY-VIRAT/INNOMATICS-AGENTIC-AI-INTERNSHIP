Task 1: Conceptual Understanding

1.What is the difference between "Love" and "love" in NLP?

In raw text, "Love" (capital L) and "love" (small l) are treated as two different tokens. This happens because computers are case-sensitive by default.

Example:
"Love is beautiful"
"I love coding"

Without preprocessing:
["Love", "love"] → treated as different words

Why it matters:
It increases vocabulary size unnecessarily
Reduces model efficiency

Solution:
Apply lowercasing
"Love" → "love"

2.What happens if stopwords are not removed?

Stopwords = common words like is, the, and, in, of

If we don’t remove them:

Problems:
Noise increases
These words don’t carry much meaning
Model becomes slower
More words = more computation
Important words get less weight

Example:

"The movie is very good"

→ "good" is important, but buried among stopwords

Result:
Reduced accuracy
Poor feature representation

3.Provide two real-world scenarios where removing stopwords can be harmful.

Scenario 1: Sentiment Analysis

Example:

"I do not like this product"
If "not" is removed:
"I like this product"

Meaning becomes completely opposite

Scenario 2: Question Answering / Search Systems

Example:

"What is the capital of India?"
Removing stopwords:
"capital India"

Loses structure and intent of the question

Key idea:

Some stopwords (like not, no, what, where) carry critical meaning

4.Explain the difference between stemming and lemmatization.

🔹 Stemming
Cuts words to root form using simple rules
May produce incorrect or non-dictionary words

Examples:

running → run

studies → studi

✔ Fast

❌ Less accurate

🔹 Lemmatization
Converts word to correct base form (lemma)
Uses dictionary + grammar rules

Examples:

running → run

better → good

✔ Accurate

❌ Slower

Task 2:
Build Advanced Preprocessing Function


```python
import re

def preprocess_text(text):
    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters (e.g., soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove special characters / punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Remove very short tokens (≤ 2) except "no", "not"
    tokens = text.split()
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            cleaned_tokens.append(word)

    # 9. Join back into sentence
    return " ".join(cleaned_tokens)

text = "WOW!!! This is soooo goooood!!! Visit http://example.com now. Email me at test@gmail.com. I have 2 dogs."
print(preprocess_text(text))
```

Task 6: Build Full Pipeline

```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = text.split()
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            cleaned_tokens.append(word)
    return " ".join(cleaned_tokens)

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for text in text_list:
        cleaned = preprocess_text(text)
        clean_sentences.append(cleaned)
        tokens = cleaned.split()
        all_tokens.extend(tokens)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

text_list = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!"
]
result = full_pipeline(text_list)
print("All Tokens:\n", result["tokens"])
print("\nClean Sentences:\n", result["clean_sentences"])
```

Task 7: Error Handling

```python
import re
from collections import Counter

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    if not text:
        return ""
    tokens = text.split()
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            cleaned_tokens.append(word)
    return " ".join(cleaned_tokens)

def full_pipeline(text_list):
    if not text_list or not isinstance(text_list, list):
        return {"tokens": [], "clean_sentences": []}
    all_tokens = []
    clean_sentences = []
    for text in text_list:
        cleaned = preprocess_text(text)
        clean_sentences.append(cleaned)
        if cleaned:
            tokens = cleaned.split()
            all_tokens.extend(tokens)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this",
    "",                     # empty string
    "😍😍😂😂🔥🔥",           # only emojis
    "1234567890",           # only numbers
    None                    # invalid input
]
result = full_pipeline(test_sentences)
print("\n--- CLEAN SENTENCES ---")
for i, sent in enumerate(result["clean_sentences"], 1):
    print(f"{i}. {sent}")
print("\n--- ALL TOKENS ---")
print(result["tokens"])
counter = Counter(result["tokens"])
print("\n--- TOP 10 MOST FREQUENT ---")
print(counter.most_common(10))
print("\n--- TOP 5 LEAST FREQUENT ---")
least_5 = sorted(counter.items(), key=lambda x: x[1])[:5]
print(least_5)
```
